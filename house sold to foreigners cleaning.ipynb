```python
departing_df = pd.read_excel('./DataSet/number_of_departing_visitors_and_average_number_of_overnights.xls')
```

```python
departing_df.head(20)
```

```python
#forwards fill
departing_df = departing_df.fillna(method='ffill')
```

```python
departing_df.head(20)
```

```python
#drop rows having year <2013
departing_df = departing_df[departing_df['Yıl - Year'] >= 2013]
```

```python
departing_df.head(20)
```

```python
#drop rows having Yıllık - Annual	in Yıllık - Annual \nÇeyrek - Quarter column
departing_df = departing_df[departing_df['Yıllık - Annual \nÇeyrek - Quarter'] != 'Yıllık - Annual']
 
```

```python
departing_df.head(20)
```

```python
#drop Vatandaş \n(yurt dışında ikamet eden) \nCitizen (Resident abroad) column
departing_df = departing_df.drop(['Vatandaş \n(yurt dışında ikamet eden) \nCitizen (Resident abroad)'], axis=1)
```

```python
departing_df
```

```python
#drop row having Yıllık(1) - Annual in Yıllık - Annual \nÇeyrek - Quarter column
departing_df = departing_df[departing_df['Yıllık - Annual \nÇeyrek - Quarter'] != 'Yıllık(1) - Annual']
```

```python
departing_df.head(20)
```

```python
# Create separate columns for each quarter
departing_df_pivot = departing_df.pivot(index='Yıl - Year', columns='Yıllık - Annual \nÇeyrek - Quarter', values='Toplam -Total')

# Rename the columns for better readability
departing_df_pivot.columns = ['Quarter I', 'Quarter II', 'Quarter III', 'Quarter IV']

# Reset the index to make 'Yıl - Year' a regular column
departing_df_pivot.reset_index(inplace=True)
```

```python
departing_df_pivot.head(20)
```

```python
#remove the 2020 row
departing_df_pivot = departing_df_pivot[departing_df_pivot['Yıl - Year'] != 2020]
```

```python
departing_df_pivot.head(20)
```

```python
#substitute 2023 quarter 4 with median of quarter 4 column
departing_df_pivot.loc[departing_df_pivot['Yıl - Year'] == 2023, 'Quarter IV'] = departing_df_pivot['Quarter IV'].median()
```

```python
departing_df_pivot.head(20)
```

```python
df1.head(20)
```

```python
merged_df = pd.merge(departing_df_pivot, df1, left_on='Yıl - Year', right_on='Yıl\nYear', how='inner')
```

```python
merged_df
```

```python
#Change the values of Quarter I , Quarter II, Quarter III, Quarter IV columns to 1000s
merged_df['Quarter I'] = merged_df['Quarter I'] / 1000
merged_df['Quarter II'] = merged_df['Quarter II'] / 1000
merged_df['Quarter III'] = merged_df['Quarter III'] / 1000
merged_df['Quarter IV'] = merged_df['Quarter IV'] / 1000
merged_df['Toplam\nTotal'] = merged_df['Toplam\nTotal'] / 1000
```

```python
merged_df
```

```python
#add a total column for Quarter 1,2,3,4
merged_df['Total_Houses'] = merged_df['Quarter 1'] + merged_df['Quarter 2'] + merged_df['Quarter 3'] + merged_df['Quarter 4']
```

```python
merged_df.columns
```

```python

# Create subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))

# Grouped bar plot for houses sold in each quarter
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter I', label='Quarter I - Houses Sold', color='blue', ax=ax1)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter II', label='Quarter II - Houses Sold', color='lightblue', ax=ax1)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter III', label='Quarter III - Houses Sold', color='skyblue', ax=ax1)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter IV', label='Quarter IV - Houses Sold', color='steelblue', ax=ax1)

# Create a secondary y-axis for total departures
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter 1', label='Quarter I - Departures', color='red', ax=ax2)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter 2', label='Quarter II - Departures', color='orange', ax=ax2)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter 3', label='Quarter III - Departures', color='yellow', ax=ax2)
sns.barplot(data=merged_df, x='Yıl - Year', y='Quarter 4', label='Quarter IV - Departures', color='green', ax=ax2)

# Set labels and title
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Houses Sold', color='blue')
ax2.set_ylabel('Total Departures / Departures per Quarter', color='black')
plt.title('Impact of Departures on Houses Sold in Each Quarter')

# Display the plot
plt.show()
```

```python
# Plotting
fig, ax1 = plt.subplots(figsize=(15, 15))

# Line plot for total houses sold in each quarter
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter I', label='Quarter I - Houses Sold', ax=ax1)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter II', label='Quarter II - Houses Sold', ax=ax1)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter III', label='Quarter III - Houses Sold', ax=ax1)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter IV', label='Quarter IV - Houses Sold', ax=ax1)

# Create a total line for houses sold on the primary y-axis
sns.lineplot(data=merged_df, x='Yıl - Year', y='Total_Houses', label='Total Houses Sold',linestyle='--', color='orange', ax=ax1)

# Create a secondary y-axis for both total and quarterly departures
ax2 = ax1.twinx()
sns.lineplot(data=merged_df, x='Yıl - Year', y='Toplam\nTotal', label='Total Departures', linestyle='--', color='black', ax=ax2)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter 1', label='Quarter 1 Departures', linestyle=':', color='red', ax=ax2)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter 2', label='Quarter 2 Departures', linestyle=':', color='green', ax=ax2)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter 3', label='Quarter 3 Departures', linestyle=':', color='blue', ax=ax2)
sns.lineplot(data=merged_df, x='Yıl - Year', y='Quarter 4', label='Quarter 4 Departures', linestyle=':', color='purple', ax=ax2)

# Set labels and title
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Houses Sold', color='blue')
ax2.set_ylabel('Departures', color='black')

# Adjust legend placement
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax2.legend(loc='upper left', bbox_to_anchor=(1, 0.65))

plt.title('Impact of Departures on Houses Sold in Each Quarter')

# Display the plot
plt.show()
```
