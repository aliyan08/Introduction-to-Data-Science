```python
df = pd.read_excel('./DataSet/arriving citizens by the purpose of visit and by the way of entrance (resident citizens inside turke.xls')
```

```python
df.head(20)
```

```python
df.columns = ['Year', 'Quarter', 'Total', 'Travel, entertainment, sportive or cultural activities', 'Visiting relatives and friends', 'Education, training (less than a year)', 'Health or medical reasons (less than a year)', 'Religion / Pilgrimage', 'Shopping', 'Business Travel','Other','Accompanying persons']
```

```python
df
```

```python
#replace 'Yıllık - Annual'  to 'Annual' in quarter column
df['Quarter'] = df['Quarter'].replace('Yıllık - Annual', 'Annual')
```

```python
df.head(20)
```

```python
#keep values having "Annual" in Quarter column and remove others
df = df[df['Quarter'] == 'Annual']
```

```python
df.head(20)
```

```python
df.info()
```

```python
#type cast Total to int
df['Total'] = df['Total'].astype(int)
```
